Store the features and target as variables X and y respectively.    
å°‡ç‰¹å¾µåŠç›®æ¨™ä»¥X, yä½œç‚ºè®Šæ•¸åˆ†åˆ¥é€²è¡Œå„²å­˜

```python
X = encoded_data.drop(columns='Improvement_in_Learning_Outcomes')
y = encoded_data['Improvement_in_Learning_Outcomes']
```

Split the data into a 7:3 ratio, with 70% for the training set and 30% for the test set.    
å°‡è³‡æ–™å€åˆ†æˆ7: 3
è¨“ç·´é›†ä½”0.7 æ¸¬è©¦é›†ä½”0.3

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=128)
```

To avoid large variations or outliers in the data, standardization is applied to modify the data values.    
ç‚ºäº†é¿å…è³‡æ–™å·®ç•°éå¤§æˆ–æ˜¯éæ–¼é›¢ç¾¤ï¼Œå› æ­¤é€éæ¨™æº–åŒ–ä¾†ä¿®æ”¹æ•¸æ“šå…§å®¹

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

Due to the large number of features (20 features), feature selection is performed using a random forest approach. The top five important features are selected as the feature data for training the model.   
ç”±æ–¼ç‰¹å¾µéå¤š(20å€‹ç‰¹å¾µ)ï¼Œå› æ­¤é€ééš¨æ©Ÿæ£®æ—çš„æ–¹å¼é€²è¡Œç¯©é¸ç‰¹å¾µ
ä¸¦ä¸”å–å‰äº”å€‹é‡è¦ç‰¹å¾µä½œç‚ºè¨“ç·´çš„æ¨¡å‹çš„ç‰¹å¾µæ•¸æ“š


```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# ä½¿ç”¨éš¨æ©Ÿæ£®æ—ç¯©é¸ç‰¹å¾µ
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_std, y_train)

# ç²å–ç‰¹å¾µé‡è¦æ€§
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]  # æŒ‰é‡è¦æ€§æ’åº
print("ç‰¹å¾µé‡è¦æ€§:", importances)
print("ç‰¹å¾µæ’åº:", indices)

# é¸æ“‡å‰äº”å€‹é‡è¦ç‰¹å¾µ
X_train_selected = X_train_std[:, indices[:5]]
X_test_selected = X_test_std[:, indices[:5]]
```

1. Using KNN   
To enable the model to automatically find the optimal parameter combination, param_grid is used to define the search space, and GridSearchCV is applied to tune the parameters and evaluate the performance of the KNN model. This process identifies the best parameters that maximize the model's accuracy.
   
ç‚ºäº†è®“æ¨¡å‹èƒ½å¤ è‡ªå‹•å°‹æ‰¾æœ€ä½³çš„åƒæ•¸çµ„åˆï¼Œå› æ­¤åˆ©ç”¨ param_grid å®šç¾©æœç´¢ç©ºé–“ä¸¦é€é GridSearchCV å° KNN æ¨¡å‹é€²è¡Œåƒæ•¸èª¿æ•´å’Œæ€§èƒ½è©•ä¼°ä»¥é¸å‡ºèƒ½æœ€å¤§åŒ–æ¨¡å‹æº–ç¢ºç‡çš„æœ€ä½³åƒæ•¸ã€‚
(1) n_neighborsï¼ˆé„°å±…æ•¸é‡ï¼‰
å«ç¾©ï¼šK è¡¨ç¤ºåœ¨ KNN æ¨¡å‹ä¸­ï¼Œåˆ¤æ–·ä¸€å€‹æ¨£æœ¬çš„åˆ†é¡æ™‚ï¼Œè€ƒæ…®æœ€è¿‘çš„ K å€‹é„°å±…ã€‚
(2) weightsï¼ˆæ¬Šé‡åƒæ•¸ï¼‰
å«ç¾©ï¼šæ±ºå®šé„°å±…å°é æ¸¬çµæœçš„è²¢ç»æ–¹å¼ã€‚
(3) metricï¼ˆè·é›¢åº¦é‡ï¼‰
å«ç¾©ï¼šè¨ˆç®—æ¨£æœ¬ä¹‹é–“è·é›¢çš„æ–¹æ³•ã€‚


```python
from sklearn.model_selection import GridSearchCV

# å®šç¾© K å€¼ç¯„åœ
param_grid = {
    'n_neighbors': range(1, 50),  # K å€¼å¾ 1 åˆ° 50
    'weights': ['uniform', 'distance'],  # æ¸¬è©¦æ¬Šé‡åƒæ•¸
    'metric': ['euclidean', 'manhattan']  # æ¸¬è©¦ä¸åŒçš„è·é›¢åº¦é‡
}
```

```python
from sklearn.neighbors import KNeighborsClassifier
# å‰µå»º KNN æ¨¡å‹
knn = KNeighborsClassifier()
```

```python
# ä½¿ç”¨ GridSearchCV
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train_selected, y_train)
```

```python
# ç²å–æœ€ä½³åƒæ•¸
print("æœ€ä½³åƒæ•¸:", grid_search.best_params_)
print("æœ€ä½³æº–ç¢ºç‡:", grid_search.best_score_)

# ä½¿ç”¨æœ€ä½³æ¨¡å‹æ¸¬è©¦
best_knn = grid_search.best_estimator_
accuracy = best_knn.score(X_test_selected, y_test)
print("æ¸¬è©¦é›†æº–ç¢ºç‡:", accuracy)
```

Using the KNN model, the best accuracy achieved is 0.51. However, since the target data only consists of two classes, (No) represented by 0 and (Yes) represented by 1, this model's performance is no better than random guessing, making it ineffective for this task.   
é€éKNNæ¨¡å‹æœ€ä½³çš„æº–ç¢ºç‡ç‚º0.51ï¼Œä½†æ­¤è³‡æ–™ç›®æ¨™åªæœ‰(No)0èˆ‡(Yes)1ï¼Œå› æ­¤æ­¤æ¨¡å‹è¨­è¨ˆçµæœæ²’ç”¨

### Use Elbow Method and Silhouette score to find the best K value (Tools)

```python
from sklearn.metrics import silhouette_score

# ä½¿ç”¨è‚˜éƒ¨æ³•å‰‡æ‰¾æœ€ä½³Kå€¼
inertias = []
silhouette_scores = []
K = range(2, 8)

for k in K:
    kmE = KMeans(n_clusters=k, random_state=77)
    kmE.fit(Xf02_scaled)
    inertias.append(kmE.inertia_)
    silhouette_scores.append(silhouette_score(Xf02_scaled, kmE.labels_))

# ç¹ªè£½è‚˜éƒ¨æ³•å‰‡åœ–
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(K, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method')

# ç¹ªè£½è¼ªå»“åˆ†æ•¸åœ–
plt.subplot(1, 2, 2)
plt.plot(K, silhouette_scores, 'rx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.tight_layout()
plt.show()

# å°å‡ºå…·é«”æ•¸å€¼
print("\nInertia values:")
for k, inertia in zip(K, inertias):
    print(f"K={k}: {inertia:.2f}")

print("\nSilhouette scores:")
for k, score in zip(K, silhouette_scores):
    print(f"K={k}: {score:.3f}")
```
